```python
df = frames[5].copy()
```

```python
df
```

```python
df['Team1'].str.split('(').apply(lambda x: x[1].strip())
```

```python
# Selecting column "Team 1", turning into str and split on '('
# Apply method and using lambda function
# stripping [1] index (which is the number),and replace ")" with space. Then convert number into a float.
# assign to a new column for Team1 rounds won.

# Team1 col. split at "(" and keeping [0](team name)

df['Round1'] = df['Team1'].str.split('(').apply(lambda x: x[1].strip()).str.replace(")", "").astype(np.float)
df['Team1'] = df['Team1'].str.split('(').apply(lambda x: x[0].strip())
```

```python
# splitting and grabing Team2 rounds won.
# creating new column

df['Round2'] = df['Team2'].str.split('(').apply(lambda x: x[1].strip()).str.replace(")", "").astype(np.float)
df['Team2'] = df['Team2'].str.split('(').apply(lambda x: x[0].strip())
```

```python
# splitting map name at space and grabbing real name
df['Map'] = df['Map'].str.split(" ").apply(lambda x: x[2])
```

```python
df.head()
```

```python
df['Win1'] = (df['Round1'] > df['Round2']).astype(np.int)
df['Win2'] = (df['Round1'] < df['Round2']).astype(np.int)

df.head()
```

```python
gp = df.groupby('Team1')[['Win1']].agg(['sum', 'count']).reset_index()
gp.head()
```

```python
gp2 = df.groupby('Team2')[['Win2']].agg(['sum', 'count']).reset_index()
gp2.head()
```
